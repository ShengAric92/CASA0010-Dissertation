## X variables

### 1. Sex and Age

```python
msoa_sexage_url = "https://github.com/ShengAric92/CASA0010_dissertation/raw/main/MSOA_data/Variable_x/sex_age_estimate/sapemsoasyoatablefinal.xlsx"
msoa_sexage = pd.read_excel(msoa_sexage_url, sheet_name=5)
```

```python
msoa_sexage_copy = msoa_sexage.copy()
msoa_sexage_copy = msoa_sexage_copy.iloc[2:]
msoa_sexage_copy.reset_index(inplace=True, drop=True)
msoa_sexage_copy.columns = msoa_sexage_copy.iloc[0]
msoa_sexage_copy = msoa_sexage_copy.iloc[1:]
msoa_sexage_copy = msoa_sexage_copy.drop(columns = ['LAD 2021 Code', 'LAD 2021 Name', 'MSOA 2021 Name'])

for col in msoa_sexage_copy.columns[1:]:
    msoa_sexage_copy[col] = pd.to_numeric(msoa_sexage_copy[col], errors='coerce')
    
msoa_sexage_copy = msoa_sexage_copy.copy()

msoa_sexage_copy.columns.values[0] = 'Areacode'

msoa_sexage_copy
```

```python
# calculate female population
female_all_col = msoa_sexage_copy.filter(like='F').columns
msoa_sexage_copy['Total_female'] = msoa_sexage_copy[female_all_col].sum(axis=1)

# calculate male population
male_all_col = msoa_sexage_copy.filter(like='M').columns
msoa_sexage_copy['Total_male'] = msoa_sexage_copy[male_all_col].sum(axis=1)

# calculate total population
msoa_sexage_copy['Total_popu'] = msoa_sexage_copy['Total_female'] + msoa_sexage_copy['Total_male']
```

```python
# calculate age 17-29 population
popu_17to29_col = msoa_sexage_copy.filter(regex='^[MF](1[7-9]|2[0-9])$').columns
msoa_sexage_copy['17to29popu'] = msoa_sexage_copy[popu_17to29_col].sum(axis=1)

# calculate age 30-39 population
popu_30to39_col = msoa_sexage_copy.filter(regex='^[MF](3[0-9])$').columns
msoa_sexage_copy['30to39popu'] = msoa_sexage_copy[popu_30to39_col].sum(axis=1)

# calculate age 40-49 population
popu_40to49_col = msoa_sexage_copy.filter(regex='^[MF](4[0-9])$').columns
msoa_sexage_copy['40to49popu'] = msoa_sexage_copy[popu_40to49_col].sum(axis=1)

# calculate age 50-59 population
popu_50to59_col = msoa_sexage_copy.filter(regex='^[MF](5[0-9])$').columns
msoa_sexage_copy['50to59popu'] = msoa_sexage_copy[popu_50to59_col].sum(axis=1)

# calculate age 60-69 population
popu_60to69_col = msoa_sexage_copy.filter(regex='^[MF](6[0-9])$').columns
msoa_sexage_copy['60to69popu'] = msoa_sexage_copy[popu_60to69_col].sum(axis=1)

# calculate age 70-79 population
popu_70to79_col = msoa_sexage_copy.filter(regex='^[MF](7[0-9])$').columns
msoa_sexage_copy['70to79popu'] = msoa_sexage_copy[popu_70to79_col].sum(axis=1)

# calculate age 80+ population
popu_80plus_col = msoa_sexage_copy.filter(regex='^[MF](8[0-9]|9[0-9]|100)$').columns
msoa_sexage_copy['80pluspopu'] = msoa_sexage_copy[popu_80plus_col].sum(axis=1)
```

```python
msoa_sexage_copy
```

```python
Sexage = msoa_sexage_copy[['Areacode', 'Total', 'Total_female',
                           '17to29popu', '30to39popu', '40to49popu', '50to59popu',
                           '60to69popu', '70to79popu', '80pluspopu']]
Sexage.columns = ['MSOA21CD', 'Total', 'Total_female', 'Age17_29', 'Age30_39', 'Age40_49', 'Age50_59',
                  'Age60_69', 'Age70_79', 'Age80']
Sexage
```

### 3. 2021MSOA code to 2011MSOA code

```python
Census21 = [Sexage, Ethicdf]
```

```python
Merge21 = Census21[0]
for df in Census21[1:]:
    Merge21 = Merge21.merge(df, on='MSOA21CD', how='left')
```

```python
# Import the assistance dataframe for 2021 and 2011 MSOA code mapping
MSOAMAP_url = "https://github.com/ShengAric92/CASA0010_dissertation/raw/main/MSOA_data/21to11code/MSOA_(2011)_to_MSOA_(2021)_to_Local_Authority_District_(2022)_Lookup_for_England_and_Wales_-1574171694243467521.csv"
MSOAMAP = pd.read_csv(MSOAMAP_url)
MSOAMAP = MSOAMAP[['MSOA11CD', 'CHGIND', 'MSOA21CD']]
MSOAMAP
```

```python
Merge = pd.merge(MSOAMAP, Merge21, how='left', left_on='MSOA21CD', right_on='MSOA21CD')

MergeU = Merge[Merge['CHGIND'] == 'U']
MergeS = Merge[Merge['CHGIND'] == 'S']
MergeM = Merge[Merge['CHGIND'] == 'M']
MergeX = Merge[Merge['CHGIND'] == 'X']
```

```python
code11_col = ['MSOA11CD'] + list(Merge.columns)[3:]
num_col = list(Merge.columns)[3:]
```

```python
Merge11U = MergeU[code11_col]
Merge11U
```

```python
Merge11S = MergeS[code11_col].groupby('MSOA11CD').sum()
Merge11S = Merge11S.reset_index()
Merge11S
```

```python
Merge11M = MergeM[code11_col].copy()
Merge11M[num_col] = Merge11M[num_col].apply(lambda x: x / 2)
Merge11M.head()
```

```python
MergeX_copy = MergeX.copy()
MergeX_copy['MSOA21CD_count'] = MergeX_copy.groupby('MSOA21CD')['MSOA21CD'].transform('count')
for col in num_col:
    MergeX_copy[col] = MergeX_copy[col] / MergeX_copy['MSOA21CD_count']
    
MergeX_copy.drop(columns=['MSOA21CD_count'], inplace=True)

Merge11X = MergeX_copy[code11_col].groupby('MSOA11CD').sum()
Merge11X = Merge11X.reset_index()
```

```python
Merge11X
```

```python
# combine all above data partition
# Now 7201 MSOA for 2011 code (England and Walse)
Merge11 = pd.concat([Merge11U, Merge11S, Merge11M, Merge11X], axis=0, ignore_index=True)
Merge11
```

### 4. Calculate rate and density (Sex, Age, Ethic, Population density)

```python
Merge11.columns
```

```python
Merge11data = Merge11.copy()
```

```python
Merge11data['asianr'] = (Merge11data.asian / Merge11data.ethicall) * 100
Merge11data['blackr'] = (Merge11data.black / Merge11data.ethicall) * 100
Merge11data['mixedr'] = (Merge11data.mixed / Merge11data.ethicall) * 100
Merge11data['whiter'] = (Merge11data.white / Merge11data.ethicall) * 100
Merge11data['otherr'] = (Merge11data.other / Merge11data.ethicall) * 100
```

```python
Merge11data['femaler'] = (Merge11data.Total_female / Merge11data.Total) * 100
Merge11data['Age17_29r'] = (Merge11data.Age17_29 / Merge11data.Total) * 100
Merge11data['Age30_39r'] = (Merge11data.Age30_39 / Merge11data.Total) * 100
Merge11data['Age40_49r'] = (Merge11data.Age40_49 / Merge11data.Total) * 100
Merge11data['Age50_59r'] = (Merge11data.Age50_59 / Merge11data.Total) * 100
Merge11data['Age60_69r'] = (Merge11data.Age60_69 / Merge11data.Total) * 100
Merge11data['Age70_79r'] = (Merge11data.Age70_79 / Merge11data.Total) * 100
Merge11data['Age80r'] = (Merge11data.Age80 / Merge11data.Total) * 100
```

```python
# Population density (per hectare)
Areadf = msoa_england_shp[['msoa11cd', 'st_areasha']].copy()
Areadf.columns = ['MSOA11CD', 'area']

# convert to hectare
Areadf['area'] = Areadf['area'] / 10000

# Merge with area
Merge11data = pd.merge(Merge11data, Areadf, how='left', left_on='MSOA11CD', right_on='MSOA11CD')

# calculate density
Merge11data['popudensity'] = Merge11data.Total / Merge11data.area
```

```python
Census_data = Merge11data[Merge11data_col]

# only with england 6791 MSOA
Census_data = Census_data[Census_data['MSOA11CD'].str.startswith("E")]
```

### 8. IMD data (LSOA to MSOA using population weighted method)

```python
# import MSOA and LSOA map data
MSOAtoLSOA_url = "https://github.com/ShengAric92/CASA0010_dissertation/raw/main/MSOA_data/Variable_x/IMD/LSOA_MSOA.csv"
MSOAtoLSOA = pd.read_csv(MSOAtoLSOA_url)
MSOAtoLSOA = MSOAtoLSOA[['MSOA Code', 'LSOA Code']]
MSOAtoLSOA.columns = ['MSOA11CD', 'LSOA11CD']
MSOAtoLSOA = MSOAtoLSOA[MSOAtoLSOA.MSOA11CD.str.startswith("E")]
```

```python
# import LSOA IMD data, with population information
IMD_LSOA_url = "https://github.com/ShengAric92/CASA0010_dissertation/raw/main/MSOA_data/Variable_x/IMD/File_7_-_All_IoD2019_Scores__Ranks__Deciles_and_Population_Denominators_3.csv"
IMD_LSOA_score = pd.read_csv(IMD_LSOA_url)
IMD_LSOA_score = IMD_LSOA_score[['LSOA code (2011)', 'Index of Multiple Deprivation (IMD) Score', 'Total population: mid 2015 (excluding prisoners)']]
IMD_LSOA_score.columns = ['LSOA11CD', 'IMDscore_LSOA', 'population']
```

```python
MSOAtoLSOA = pd.merge(MSOAtoLSOA, IMD_LSOA_score, how='left', left_on='LSOA11CD', right_on='LSOA11CD')
MSOAIMD = MSOAtoLSOA.copy()
```

```python
# population weighted IMD
MSOAIMD['weighted_IMD'] = MSOAIMD['population'] * MSOAIMD['IMDscore_LSOA']

MSOAIMD = MSOAIMD.groupby('MSOA11CD').agg(
    total_population=('population', 'sum'),
    total_weighted_IMD=('weighted_IMD', 'sum')
)

MSOAIMD['IMDscore'] = MSOAIMD['total_weighted_IMD'] / MSOAIMD['total_population']

# Ranking the MSOA IMD socre
MSOAIMD['IMDrank'] = MSOAIMD['IMDscore'].rank(ascending=False, method='min')

# according rank to calculate decile
MSOAIMD['IMDdecile'] = pd.qcut(MSOAIMD['IMDrank'], 10, labels=False) + 1

# rank to quintile
MSOAIMD['IMDquintile'] = pd.qcut(MSOAIMD['IMDrank'], 5, labels=False) + 1

MSOAIMD = MSOAIMD.reset_index()[['MSOA11CD', 'IMDscore', 'IMDrank', 'IMDdecile', 'IMDquintile']]
```
